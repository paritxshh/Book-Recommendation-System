```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# reading files
data = pd.read_csv('dataset.csv', encoding = 'latin-1')
books = pd.read_csv('ratings.csv', encoding = 'latin-1')
```

```python
# Checking the data using head function
books.head()
```

```python
# calculating the mean
rating = pd.DataFrame(books.groupby('BOOK NUMBER')['no_of_ratings'].mean())
rating.head()
```

```python
# getting the description of rating
rating.describe()
```

```python
# sorting based on no of ratings that each book got
rating.sort_values('no_of_ratings', ascending=False).head()
```

```python
# Preparing data table for analysis
ratings_pivot = pd.pivot_table(data=books, values='user_rtg', index='user_id', columns='BOOK NUMBER')
ratings_pivot.head()
```

```python
# getting the users who rated this particular book (most rated) and making sure rating is not zero
LifeofPi_rating = ratings_pivot[31]
LifeofPi_rating[LifeofPi_rating>=0]
```

```python
# finidng similar books to One Man Out book using Pearson correlation
similar_to_LifeofPi = ratings_pivot.corrwith(LifeofPi_rating)
corr_LifeofPi = pd.DataFrame(similar_to_LifeofPi, columns=['PearsonR'])
corr_LifeofPi.dropna(inplace=True)
corr_LifeofPi.head()
```

```python
# getting the details for most similar books
book_corr_LifeofPi = pd.DataFrame([26, 49, 55, 71, 90, 17, 81, 76, 13, 31], 
                      index = np.arange(10), columns=['BOOK NUMBER'])
summary = pd.merge(book_corr_LifeofPi, data,on='BOOK NUMBER')
summary
```
