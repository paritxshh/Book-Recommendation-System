```python
# Getting recommendation based on No. Of ratings
rating_count = pd.DataFrame(rating, columns=['BOOK NUMBER','no_of_ratings'])
# Sorting and dropping the duplicates
rating_count.sort_values('no_of_ratings', ascending=False).drop_duplicates().head(10)
```

```python
# getting the detail of 5 most rated books
most_rated_books = pd.DataFrame([97, 85, 32, 66, 12], index=np.arange(5), columns=['BOOK NUMBER'])
detail = pd.merge(most_rated_books, book, on='BOOK NUMBER')
detail
```

```python
# getting the most rated book
most_rated_book = pd.DataFrame(rating, columns=['BOOK NUMBER', 'user_id', 'avg_rating', 'no_of_ratings'])
most_rated_book.max()
```
